# PCA of IRIS dataset results

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
from sklearn.decomposition import PCA
```

```python
X_iris=pd.read_csv("./iris.csv",sep=',',header=None,usecols=[0,1,2,3])
Y_iris=pd.read_csv("./iris.csv",sep=',',header=None,usecols=[4])
Y_iris=Y_iris.values.flatten()
target_names=np.unique(Y_iris)
target_names
```

```python
pca=PCA(X_iris.shape[1])
X_iris_r=pca.fit(X_iris).transform(X_iris)
y_iris=np.array([])
for i in Y_iris:
    if i=="Iris-setosa":
        y_iris=np.append(y_iris,0)
    if i=="Iris-versicolor":
        y_iris=np.append(y_iris,1)
    if i=="Iris-virginica":
        y_iris=np.append(y_iris,2)
plt.figure()

colors=['navy', 'turquoise','darkorange']
lw=2

for color, i,target_name in zip(colors,[0,1,2],target_names):
    plt.scatter(X_iris_r[y_iris==i,0],X_iris_r[y_iris==i,1],color=color,alpha=.8, lw=lw,
               label=target_name)
plt.legend(loc='best',shadow=False, scatterpoints=1)
plt.title('PCA of IRIS dataset 2D')
plt.show()
```

```python
for color, i,target_name in zip(colors,[0,1,2],target_names):
    plt.scatter(X_iris_r[y_iris==i,0],X_iris_r[y_iris==i,1],color=color,alpha=.8, lw=lw,
           label=target_name)
plt.legend(loc='best',shadow=False, scatterpoints=1)
plt.title('PCA of IRIS dataset 2D')
plt.show()
```

```python
fig=plt.figure(figsize=(12,8))
ax=plt.axes(projection='3d')
colors=[]
for v in y_iris:
    if v==0:
        colors.append('#000080')
    elif v==1:
        colors.append('#40e0d0')
    else:
        colors.append('#ff8c00') 
sctt=ax.scatter3D(X_iris_r[:,0],X_iris_r[:,1],X_iris_r[:,2],c=colors, s=50, alpha=0.8)
plt.title("PCA of IRIS dataset 3D")
```

```python
n, bins, patches = plt.hist(y_iris, bins=50, facecolor='#D62AD0', edgecolor='#e0e0e0', linewidth=0.5, alpha=1)
```

# PCA of HEART dataset results

```python
X_heart=pd.read_csv("./heart_processed.txt",sep=' ',header=None)
X_heart=X_heart.loc[:,X_heart.columns!=13]
Y_heart=pd.read_csv("./heart_processed.txt",sep=' ',header=None,usecols=[13])
Y_heart=Y_heart.values.flatten()
target_names=np.unique(Y_heart)
target_names=np.where(target_names==0,'Not disease',target_names)
target_names=np.where(target_names=='1','Disease',target_names)
target_names
```

```python
pca=PCA(X_heart.shape[1])
X_heart_r=pca.fit(X_heart).transform(X_heart)
y_heart=Y_heart
plt.figure(figsize=(10,10))

colors=['green', 'red']
lw=2

for color, i,target_name in zip(colors,[0,1],target_names):
    plt.scatter(X_heart_r[y_heart==i,0],X_heart_r[y_heart==i,1],color=color,alpha=.8, lw=lw,
               label=target_name)
plt.legend(loc='best',shadow=False, scatterpoints=1)
plt.title('PCA of heart dataset 2D')
plt.show()
```

```python
fig=plt.figure(figsize=(12,8))
ax=plt.axes(projection='3d')
colors=[]
for v in y_heart:
    if v==0:
        colors.append('#FF0000')
    elif v==1:
        colors.append('#00FF00')
sctt=ax.scatter3D(X_heart_r[:,0],X_heart_r[:,1],X_heart_r[:,2],c=colors, s=50, alpha=0.8)
plt.title("PCA of Heart dataset 3D")
```

```python
n, bins, patches = plt.hist(y_heart, bins=50, facecolor='#D62AD0', edgecolor='#e0e0e0', linewidth=0.5, alpha=1)
```
