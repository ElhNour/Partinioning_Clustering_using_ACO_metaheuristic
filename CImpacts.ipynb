```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from pso import ParticleSwarmOptimizedClustering
from utils import normalize
```

```python
data = pd.read_csv('./data/iris.data', sep=',', header=None)
x = data.drop([4], axis=1)
x = x.values
x = normalize(x)
```

```python
test_iteration = 20
nb_iter = 2000
dump_w= 0.99
C1 = 0
C2 = 0
incrc1 = 2/test_iteration
bests = []
for i in range(test_iteration):
    print(f" C1={C1},C2={C2}")
    pso = ParticleSwarmOptimizedClustering(
        n_cluster=3, n_particles=20,dump_w=dump_w,c1=C1,c2=C2, data=x, hybrid=True, max_iter=nb_iter, print_debug=nb_iter)
    pso.run()
    bests.append([C1,pso.gbest_score])
    C1+=incrc1  

df = pd.DataFrame(bests, columns =['C1', 'Gbest score'])
  
print(df) 
df.to_excel('./tests/impact_c1_iris_res.xlsx', index=False)
    
```
